```python
import pandas as pd
pd.plotting.register_matplotlib_converters()
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
print("Setup Complete")
```

```python
museum_filepath = 'gdrive/My Drive/Museum_Visitors.csv'
museum_data = pd.read_csv(museum_filepath, index_col= "Month", parse_dates= True)
museum_data
```

```python
museum_data.head()
```

```python
museum_data.tail()
```

```python
sns.lineplot(data=museum_data)
```

```python
# Set the width and height of the figure
plt.figure(figsize=(14,6))
# Line chart showing the number of visitors to each museum over time
sns.lineplot(data=museum_data)
# Add title
plt.title("Monthly Visitors to Los Angeles City Museums")
```

```python
list(museum_data.columns)
```

```python
# Set the width and height of the figure
plt.figure(figsize=(12,6))
# Add title
plt.title("Monthly Visitors to Chinese American Museum")
# Line chart showing the number of visitors to Avila Adobe over time
sns.lineplot(data=museum_data['Chinese American Museum'], label='Chinese American Museum')
# Add label for horizontal axis
plt.xlabel("Date")
plt.ylabel("number of Visitors")
```

```python
# Set the width and height of the figure
plt.figure(figsize=(12,6))
# Add title
plt.title("Monthly Visitors to Avila Adobe")
# Line chart showing the number of visitors to Avila Adobe over time
sns.lineplot(data=museum_data['Avila Adobe'], label='Avila Adobe')
# Add label for horizontal axis
plt.xlabel("Date")
plt.ylabel("number of Visitors")
```
